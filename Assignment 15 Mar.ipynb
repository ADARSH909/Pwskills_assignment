```python
Q2- What is supervised learning? List some examples of supervised learning
Ans:
    Supervised learning is a type of machine learning in which the algorithm is trained on labeled data. In other words, the input data is accompanied 
    by corresponding output data, or "labels," that the algorithm uses to learn how to make predictions or decisions. The goal of supervised learning 
    is to create a model that can accurately predict or classify new data.

Here are some examples of supervised learning:

    Image classification: Given a set of labeled images, the algorithm learns to classify new images into different categories. For example, an image 
    classification algorithm could be trained to identify different species of animals or recognize handwritten digits.

    Sentiment analysis: Given a set of labeled text data, the algorithm learns to classify new text data as either positive or negative. For example, 
    a sentiment analysis algorithm could be trained to identify whether a customer review of a product is positive or negative.

    Fraud detection: Given a set of labeled financial transactions, the algorithm learns to identify new transactions that are likely to be fraudulent.
    For example, a fraud detection algorithm could be trained to detect unusual patterns of activity in credit card transactions.

    Medical diagnosis: Given a set of labeled medical images or patient data, the algorithm learns to diagnose new patients based on their symptoms or
    medical history. For example, a medical diagnosis algorithm could be trained to identify different types of cancer or predict the likelihood of a
    heart attack.

    Language translation: Given a set of labeled text data in one language, the algorithm learns to translate new text data into another language. For
    example, a language translation algorithm could be trained to translate English sentences into French.

```

```python
Q3- What is supervised learning? List some examples of supervised learning?
Ans:
    Unsupervised learning is a type of machine learning in which the algorithm is trained on unlabeled data. In other words, the input data is not 
    accompanied by corresponding output data, or "labels," and the algorithm must discover patterns and relationships in the data on its own. The 
    goal of unsupervised learning is often to group or cluster data into meaningful categories or to identify patterns that can be used for further
    analysis.

Here are some examples of unsupervised learning:

    Clustering: Given a set of unlabeled data, the algorithm groups the data into clusters based on similarities in the data. For example, an 
    unsupervised clustering algorithm could group customers into different segments based on their purchase histories or group news articles into 
    different topics based on their content.

    Anomaly detection: Given a set of unlabeled data, the algorithm identifies data points that are significantly different from the rest of the data.
    For example, an unsupervised anomaly detection algorithm could identify credit card transactions that are significantly different from a customer's
    usual spending patterns.

    Dimensionality reduction: Given a set of high-dimensional data, the algorithm identifies a lower-dimensional representation of the data that 
    captures the most important features. For example, an unsupervised dimensionality reduction algorithm could be used to visualize high-dimensional 
    data in two or three dimensions.

    Association rule learning: Given a set of transaction data, the algorithm identifies rules that describe patterns in the data. For example, an 
    unsupervised association rule learning algorithm could identify that customers who buy bread are also likely to buy milk.

    Generative modeling: Given a set of data, the algorithm learns to generate new data that is similar to the original data. For example, an
    unsupervised generative modeling algorithm could generate new images of faces or new sentences in a natural language.
```

```python
What are the main differences between supervised, unsupervised, and semi-supervised learning?
Ans:
        Supervised learning: In supervised learning, the algorithm is trained on a labeled dataset, which means that the input data is already
        categorized into different classes or categories, and the algorithm is trained to predict the correct label for new, unseen data. The 
        labeled dataset acts as a teacher for the algorithm, guiding it towards the correct prediction. Examples of supervised learning include 
        image classification, speech recognition, and text categorization.

    Unsupervised learning: In unsupervised learning, the algorithm is trained on an unlabeled dataset, meaning that the input data is not categorized 
    or labeled in any way. The algorithm must find patterns or structure in the data without any external guidance. Unsupervised learning is used for
    tasks such as clustering, anomaly detection, and dimensionality reduction.

    Semi-supervised learning: Semi-supervised learning is a combination of supervised and unsupervised learning. It uses both labeled and unlabeled
    data to train the algorithm. The labeled data provides some guidance to the algorithm, while the unlabeled data helps to improve the accuracy of
    the model by allowing it to learn more about the underlying structure of the data. Semi-supervised learning is used when obtaining labeled data 
    is expensive or time-consuming, but unlabeled data is readily available.
```

```python
 What is train, test and validation split? Explain the importance of each term?
Ans:
    The training set is the data used to train the model. The algorithm uses the training set to learn the patterns and relationships in the data
    and to adjust its parameters to optimize its performance.

The validation set is used to evaluate the performance of the model during training. It is used to test the model after each training iteration, 
and to make sure that the model is not overfitting to the training data. Overfitting occurs when the model is too closely fitted to the training 
data, and performs poorly on new, unseen data.

The test set is used to evaluate the final performance of the model after it has been trained and validated. It is used to estimate the generalization
performance of the model, that is, how well it performs on new, unseen data.

The importance of each term:

    Training set: The training set is the most important part of the dataset, as it is used to train the model. The more representative the training
    set is of the problem at hand, the more accurate the model will be.

    Validation set: The validation set is important for monitoring the training process and avoiding overfitting. By evaluating the model on the
    validation set, we can make sure that the model is not too closely fitted to the training data, and that it will perform well on new, unseen data.

    Test set: The test set is important for evaluating the final performance of the model, and for estimating the generalization performance of the
    model. It provides an unbiased estimate of the performance of the model on new, unseen data.
```

```python
 How can unsupervisKd learning be used in anomaly detection?
Ans:
    Unsupervised learning can be used for anomaly detection by identifying data points that do not fit the normal pattern or distribution of the data.
    This can be done using clustering or density-based techniques.

In clustering, the algorithm groups data points that are similar to each other based on their features. Anomalies can be detected as data points that
do not belong to any cluster, or that are significantly far from their closest cluster. For example, if we have a dataset of customer transactions,
we can use clustering to group similar transactions, and then identify transactions that do not fit any cluster or are significantly far from their 
closest cluster as potential anomalies.

In density-based techniques, the algorithm identifies regions of high density in the data and considers data points that fall outside those regions as
anomalies. One example of a density-based technique is Local Outlier Factor (LOF), which identifies data points that have a lower density than their 
neighbors.

Unsupervised anomaly detection is useful when there is little or no prior knowledge about the anomalies, or when the anomalies are rare and difficult
to label. However, unsupervised anomaly detection may have higher false positive rates compared to supervised methods since it has no prior knowledge
of the anomalies. Therefore, it is important to carefully choose appropriate techniques and validate the results.
```
