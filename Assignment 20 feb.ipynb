```python
 Explain GET and POST methods?
Ans:
GET method:
The GET method is used to request data from a server. It sends a request to retrieve a resource identified by a URL. 
The request parameters, if any, are appended to the URL as a query string. The server retrieves the requested resource
and sends it back to the client. The response may contain additional headers and metadata related to the requested resource. 
The GET method is safe, meaning it should not modify the state of the server, and can be cached by intermediary servers such as web proxies.

For example, when you enter a URL into your web browser to request a webpage, your browser is making a GET request to the server hosting that page.

POST method:
The POST method is used to send data to a server to create or update a resource. Unlike the GET method, the data is not appended to 
the URL, but is included in the request body. This method can be used to submit forms, upload files, and perform other operations that
modify the state of the server. The POST method is not safe, as it can modify the state of the server and is not recommended to be cached
by intermediary servers.

For example, when you fill out a form on a website and submit it, your browser is making a POST request to the server to send the data you
entered in the form.
```

```python
Q3. Why is redirect() used in Flask?
Ans:
In Flask, the redirect() function is used to redirect the user to a different URL. It returns a redirect response with a 302 HTTP status code,
telling the user's browser to go to the new URL.

The redirect() function is commonly used in Flask applications to implement the Post/Redirect/Get (PRG) pattern. This pattern is used to avoid
the resubmission of data when the user refreshes the page after submitting a form.

Here's how the PRG pattern works:

    The user submits a form to the server using the HTTP POST method.
    The server processes the form data and sends a response with a redirect to a new URL using the redirect() function.
    The user's browser follows the redirect and sends a new HTTP GET request to the new URL.
    The server sends a response with the updated data to the user's browser.

By using the redirect() function in this way, the server ensures that the user's browser will not resubmit the form data if the user refreshes
the page. This helps to prevent duplicate submissions and other related issues.
```

```python
Q4. What are templates in Flask? Why is the render_template() function used?
Ans:
In Flask, templates are used to generate HTML responses dynamically. A template is a file that contains a mix of 
static content (HTML, CSS, etc.) and dynamic content (variables, expressions, loops, conditionals, etc.). 
Flask uses the Jinja2 templating engine to render templates and generate HTML responses.

Templates are important in Flask because they allow developers to separate the presentation layer (HTML) from the application logic. 
This makes it easier to maintain and update the application over time, as changes to the presentation layer can be made independently
of the application logic.

The render_template() function is used in Flask to render a Jinja2 template and generate an HTML response. It takes the name of the
template as an argument and any variables that should be passed to the template for rendering. The function looks for the specified
template in the application's template directory and returns the rendered HTML as a response to the user's browser.
```
