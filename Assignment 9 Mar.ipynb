```python
What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.
Ans:
Both Probability Mass Function (PMF) and Probability Density Function (PDF) are mathematical functions used to describe the
probability distribution of a random variable.

The PMF is used for discrete random variables, while the PDF is used for continuous random variables.

The PMF assigns a probability value to each possible outcome of a discrete random variable. For example, let's consider the roll of a fair 
six-sided die. The PMF of this random variable would be:
Outcome	1	2	3	4	5	6
PMF	1/6	1/6	1/6	1/6	1/6	1/6

This means that the probability of rolling a 1 is 1/6, the probability of rolling a 2 is 1/6, and so on.

The PDF, on the other hand, assigns a probability density value to each possible value of a continuous random variable. Unlike the PMF, 
the PDF does not give the probability of a specific value occurring, but rather the probability of a value falling within a certain range. 
For example, let's consider the height of adult males in a certain population. The PDF of this random variable would be a smooth curve that 
shows the probability density of different height values. Here is an example of what the PDF might look like:

PDF example

In this example, the x-axis represents height, and the y-axis represents the probability density. The curve shows that the probability density
is highest around a certain height value (the mean of the distribution), and decreases as we move further away from that value.
```

```python
What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
Ans:
The Cumulative Density Function (CDF) is a mathematical function that describes the cumulative probability distribution of a random variable, 
whether it is continuous or discrete. It is defined as the probability that a random variable X takes on a value less than or equal to a certain 
threshold value x.

Formally, for a random variable X, the CDF is defined as:

F(x) = P(X <= x)

where F(x) represents the probability that X is less than or equal to x.

For example, let's consider a simple example of the CDF of a discrete random variable. Suppose we roll a fair six-sided die. The CDF for this 
random variable would be:
Outcome	1	2	3	4	5	6
PMF	1/6	1/6	1/6	1/6	1/6	1/6
CDF	1/6	1/3	1/2	2/3	5/6	1

This table shows both the PMF and the CDF of the random variable. The PMF assigns a probability value to each outcome, while the CDF shows the
probability of getting a value less than or equal to a certain threshold. For example, the CDF value at x=3 is 1/2, meaning that the probability 
of getting a value less than or equal to 3 is 1/2.

In general, the CDF is used to provide a more complete description of the distribution of a random variable than just the PMF or PDF. It can be used
to calculate various statistical properties of the distribution, such as the mean, variance, and percentiles. Additionally, it can be used to compare
different distributions and to perform hypothesis testing.
```

```python
What is the z score? State the importance of the z score?
Ans:
    The z-score, also known as the standard score, is a measure of how many standard deviations a given observation or value is away from the mean
    of a distribution. The z-score is calculated as:

z = (x - μ) / σ

where x is the observation we want to calculate the z-score for, μ is the mean of the distribution, and σ is the standard deviation of the distribution.

The z-score is important because it allows us to compare values from different distributions on a common scale. By converting values to their 
corresponding z-scores, we can compare how far away they are from their respective means in terms of standard deviations. This is particularly 
useful when working with distributions that have different means and standard deviations, or when comparing values from different populations.

The z-score is also used in hypothesis testing, where it is used to determine the probability (p-value) of observing a certain value or more extreme 
values under a given null hypothesis. By calculating the z-score for a sample and comparing it to the z-scores corresponding to a certain level of 
significance (alpha), we can determine whether to reject or fail to reject the null hypothesis.
```

```python
What is Central Limit Theorem? State the significance of the Central Limit Theorem?
Ans:
     The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that the distribution of the means of a large number of 
    independent and identically distributed (i.i.d.) random variables will be approximately normal, regardless of the underlying distribution of 
    the random variables. In other words, if we take a large sample size from any population and compute the sample mean, the distribution of the 
    sample means will approach a normal distribution, regardless of the shape of the population distribution.

The CLT has significant importance in statistics for several reasons:

    It provides a way to estimate population parameters: The CLT allows us to use the sample mean and standard deviation to estimate population 
    parameters like the mean and standard deviation, even when we don't know the underlying population distribution.

    It is the basis for statistical inference: The CLT is the basis for many statistical inference techniques, such as confidence intervals and
    hypothesis testing. It allows us to use the sample mean to make inferences about the population mean, even when the population distribution is 
    unknown or non-normal.

    It explains why normal distribution is so common in nature: Many real-world phenomena follow a normal distribution, and the CLT explains why.
    It shows that the mean of any random variable is the result of many small, independent factors, which leads to the normal distribution.

    It is widely used in quality control and process improvement: The CLT is widely used in quality control and process improvement to monitor 
    and improve the quality of products and services. By measuring the sample mean and standard deviation, we can estimate the true mean and 
    standard deviation of a process and make decisions about whether it is operating correctly or needs improvement.
```

```python
State the assumptions of the Central Limit Theorem?
Ans:
   The Central Limit Theorem (CLT) has some assumptions that must be met in order for it to hold:

    Independence: The sample observations must be independent of each other. In other words, the value of one observation should not influence the
    value of another observation.

    Sample size: The sample size must be large enough. A rule of thumb is that the sample size should be at least 30.

    Identically distributed: The sample observations must be identically distributed. This means that they should come from the same population and
    have the same distribution.

    Finite variance: The population variance must be finite. If the population variance is infinite, the CLT may not hold.

If these assumptions are met, the distribution of the sample means will approach a normal distribution, regardless of the shape of the population
distribution. The CLT is a powerful tool in statistical inference, as it allows us to make inferences about population parameters using sample 
statistics. However, it's important to remember that the assumptions of the CLT must be met in order for it to be applicable.
```
